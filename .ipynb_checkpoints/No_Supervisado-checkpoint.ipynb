```python
#Importar las librerias para los graficos:
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
```

```python
#Importar numpy python
import numpy as np
```

```python
#Importar el sample generator de scikit-learn - make blobs
from sklearn.datasets.samples_generator import make_blobs
```

```python
#Graficarlo
X,Y = make_blobs(n_samples=300, centers=4)

plt.scatter(X[:,0], X[:,1], s=50)
```

```python
# K-means de scikit-learn
from sklearn.cluster import KMeans
```

```python
#K-means
kmeans = KMeans(n_clusters=5)
```

```python
#Entrenar algoritmo
kmeans.fit(X)
```

```python
#Clusters labels
y_means = kmeans.predict(X)
```

```python
#Centroid values
centers = kmeans.cluster_centers_
print(centers)
```

```python
#Graficar valores conjuntos
plt.scatter(X[:,0], X[:,1], c=y_means, cmap='viridis')

#Incluidos los centroides
plt.scatter(centers[:,0], centers[:,1], c='black', s=200, alpha=0.5);
```
